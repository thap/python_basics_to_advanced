```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from bokeh.plotting import figure
from bokeh.io import output_file, show
```

```python
lit = pd.read_csv('literacy_birth_rate.csv')
lit
```

```python
lit = lit.drop(lit.index[162:])
lit
```

```python
lit[['fertility']] = lit[['fertility']].astype('float')
lit[['female literacy']] = lit[['female literacy']].astype('float')
lit.info()
```

```python
p = figure(x_axis_label='fertility (children per woman)', 
           y_axis_label='female literacy (% population)')
p.circle(lit['fertility'], lit['female literacy'])
output_file('fert_lit.html', p)
show(p)
```

```python
lit_lat = lit[lit['Continent'] == 'LAT']
lit_lat.head()
```

```python
lit_af = lit[lit['Continent'] == 'AF']
lit_af.head()
```

```python
p = figure(x_axis_label='fertility (children per woman)', 
           y_axis_label='female literacy (% population)')
p.circle(lit_lat['fertility'], lit_lat['female literacy'])
p.x(lit_af['fertility'], lit_af['female literacy'])
output_file('fert_lit_separate.html', p)
show(p)
```

```python
p = figure(x_axis_label='fertility (children per woman)', 
           y_axis_label='female literacy (% population)')
p.circle(lit_lat['fertility'], lit_lat['female literacy'], size=10, alpha=0.8, 
         color='blue')
p.circle(lit_af['fertility'], lit_af['female literacy'], size=10, alpha=0.8, color='red')
output_file('fert_lit_separate_colors.html', p)
show(p)
```

```python
aapl = pd.read_csv('aapl.csv', index_col=0)
aapl.head()
```

```python
aapl['date']=pd.to_datetime(aapl['date'])
aapl.info()
```

```python
aapl = aapl.set_index('date')
aapl.head()
```

```python
p = figure(x_axis_type='datetime', x_axis_label='Date', y_axis_label='US Dollars')
p.line(aapl.index, aapl.adj_close)
output_file('line.html')
show(p)
```
