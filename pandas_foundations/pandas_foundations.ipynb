```python
sunlight = pd.read_csv('NOAA_QCLCD_2011_hourly_13904.txt', header=None)
sunlight.head()
```

```python
cols = ['Wban', 'date', 'Time', 'StationType', 'sky_condition', 'sky_conditionFlag', 'visibility', 'visibilityFlag', 'wx_and_obst_to_vision', 'wx_and_obst_to_visionFlag', 'dry_bulb_faren', 'dry_bulb_farenFlag', 'dry_bulb_cel', 'dry_bulb_celFlag', 'wet_bulb_faren', 'wet_bulb_farenFlag', 'wet_bulb_cel', 'wet_bulb_celFlag', 'dew_point_faren', 'dew_point_farenFlag', 'dew_point_cel', 'dew_point_celFlag', 'relative_humidity', 'relative_humidityFlag', 'wind_speed', 'wind_speedFlag', 'wind_direction', 'wind_directionFlag', 'value_for_wind_character', 'value_for_wind_characterFlag', 'station_pressure', 'station_pressureFlag', 'pressure_tendency', 'pressure_tendencyFlag', 'presschange', 'presschangeFlag', 'sea_level_pressure', 'sea_level_pressureFlag', 'record_type', 'hourly_precip', 'hourly_precipFlag', 'altimeter', 'altimeterFlag', 'junk']
sunlight.columns = cols
sunlight.head()
```

```python
cols_to_drop = ['sky_conditionFlag', 'visibilityFlag', 'wx_and_obst_to_vision', 'wx_and_obst_to_visionFlag', 'dry_bulb_farenFlag', 'dry_bulb_celFlag', 'wet_bulb_farenFlag', 'wet_bulb_celFlag', 'dew_point_farenFlag', 'dew_point_celFlag', 'relative_humidityFlag', 'wind_speedFlag', 'wind_directionFlag', 'value_for_wind_character', 'value_for_wind_characterFlag', 'station_pressureFlag', 'pressure_tendencyFlag', 'pressure_tendency', 'presschange', 'presschangeFlag', 'sea_level_pressureFlag', 'hourly_precip', 'hourly_precipFlag', 'altimeter', 'record_type', 'altimeterFlag', 'junk']
sunlight_d = sunlight.drop(cols_to_drop, axis='columns')
sunlight_d.head()
```

```python
sunlight_d['date'] = sunlight_d['date'].astype(str)
sunlight_d['Time'] = sunlight_d['Time'].apply(lambda x: '{:0>4}'.format(x))
date_str = sunlight_d['date'] + sunlight_d['Time']
date_times = pd.to_datetime(date_str)
sunlight_c = sunlight_d.set_index(date_times)
sunlight_c.head()
```

```python
sunlight_c.info()
```

```python
sunlight_c.loc['2011-06-20 08:00:00':'2011-06-20 09:00:00', 'dry_bulb_faren']
```

```python
sunlight_c['dry_bulb_faren'] = pd.to_numeric(sunlight_c['dry_bulb_faren'], errors='coerce')
```

```python
sunlight_c.loc['2011-06-20 08:00:00':'2011-06-20 09:00:00', 'dry_bulb_faren']
```

```python
sunlight_c['wind_speed'] = pd.to_numeric(sunlight_c['wind_speed'], errors='coerce')
sunlight_c['dew_point_faren'] = pd.to_numeric(sunlight_c['dew_point_faren'], errors='coerce')
```

```python
print(sunlight_c['dry_bulb_faren'].median())
```

```python
sunlight_c.loc['2011-Apr':'2011-Jun', 'dry_bulb_faren'].median()
```

```python
sunlight_c.loc['2011-Jan', 'dry_bulb_faren'].median()
```

```python
sunlight_c.head()
```

```python
daily_sunlight = sunlight_c.resample('D').mean()
daily_sunlight.head()
```

```python
sunny = sunlight_c.loc[sunlight_c['sky_condition'] == 'CLR']
# Resample sunny by day then calculate the max
sunny_daily_max = sunny.resample('D').max()
sunny_daily_max.head()
```

```python
overcast = sunlight_c.loc[sunlight_c['sky_condition'].str.contains('OVC')]
ovc_daily_max = overcast.resample('D').max()
ovc_daily_max.head()
```

```python
sunlight.set_index(date_times, inplace=True)
weekly_mean = sunlight[['visibility', 'dry_bulb_faren']].resample('W').mean()
print(weekly_mean.corr())
```
