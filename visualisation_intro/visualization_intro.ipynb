```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
majors = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
majors.head()
```

```python
majors.plot('Year', 'Physical Sciences', color='blue')
plt.plot('Year', 'Computer Science', data=majors, color='red')
```

```python
plt.axes([0.05, 0.05, 0.425, 0.9])
plt.plot('Year', 'Physical Sciences', data=majors, color='blue')
plt.axes([0.525, 0.05, 0.425, 0.9])
plt.plot('Year', 'Computer Science', data=majors, color='red')
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot('Year', 'Physical Sciences', data=majors, color='blue')
plt.title('Physical Sciences')

plt.subplot(1, 2, 2)
plt.plot('Year', 'Computer Science', data=majors, color='red')
plt.title('Computer Science')

plt.tight_layout()
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot('Year', 'Physical Sciences', data=majors, color='blue')
plt.title('Physical Sciences')

plt.subplot(2, 2, 2)
plt.plot('Year', 'Computer Science', data=majors, color='red')
plt.title('Computer Science')

plt.subplot(2, 2, 3)
plt.plot('Year', 'Health Professions', data=majors, color='green')
plt.title('Health Professions')

plt.subplot(2, 2, 4)
plt.plot('Year', 'Education', data=majors, color='cyan')
plt.title('Education')

plt.tight_layout()
plt.show()
```

```python
plt.plot('Year', 'Computer Science', data=majors, color='red')
plt.plot('Year', 'Physical Sciences', data=majors, color='blue')
plt.xlabel('Year')
plt.ylabel('Women Percentage (%)')
plt.xlim(1990, 2010)
plt.ylim(0, 50)
plt.title('Degrees Awarded to Women (1990-2010)')
plt.legend()
plt.show()
```

```python
plt.plot('Year', 'Computer Science', data=majors, color='red')
plt.plot('Year', 'Physical Sciences', data=majors, color='blue')
plt.axis((1990, 2010, 0, 50))
plt.show()
```

```python
plt.plot('Year', 'Computer Science', data=majors, \
         color='red', label='Computer Science')
plt.plot('Year', 'Physical Sciences', data=majors, \
         color='blue', label='Physical Sciences')
plt.legend(loc='lower center')
plt.xlabel('Year')
plt.ylabel('Women Percentage (%)')
plt.title('Undergradaduate Enrollment of Women')
plt.show()
```

```python
plt.plot('Year', 'Computer Science', data=majors, \
         color='red', label='Computer Science')
plt.plot('Year', 'Physical Sciences', data=majors, \
         color='blue', label='Physical Sciences')
plt.legend(loc='lower center')
cs_max = majors['Computer Science'].max()
yr_max = majors.loc[majors['Computer Science'].idxmax(), 'Year']
plt.annotate('Maximum', xy=(yr_max, cs_max), xytext=(yr_max+5, cs_max+5), \
             arrowprops=dict(facecolor='black'))
plt.xlabel('Year')
plt.ylabel('Women Percentage (%)')
plt.title('Endergraduate Enrollment of Women')
plt.show()
```

```python
plt.style.use('ggplot')
plt.subplot(2, 2, 1)
plt.plot('Year', 'Physical Sciences', data=majors, color='blue')
plt.title('Physical Sciences')
plt.subplot(2, 2, 2)
plt.plot('Year', 'Computer Science', data=majors, color='red')
plt.title('Computer Science')
cs_max = majors['Computer Science'].max()
yr_max = majors.loc[majors['Computer Science'].idxmax(), 'Year']
plt.annotate('Maximum', xy=(yr_max, cs_max), xytext = (yr_max-1, cs_max-10), \
             arrowprops=dict(facecolor='black'))
plt.subplot(2, 2, 3)
plt.plot('Year', 'Health Professions', data=majors, color='green')
plt.title('Health Professions')
plt.subplot(2, 2, 4)
plt.plot('Year', 'Education', data=majors, color='cyan')
plt.title('Education')
plt.tight_layout()
plt.show()
```

```python
u = np.linspace(-2, 2, 41)
v = np.linspace(-1, 1, 21)

X, Y = np.meshgrid(u, v)
Z = np.sin(3*np.sqrt(X**2 + Y**2))

plt.pcolor(Z)
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.contour(X, Y, Z)

plt.subplot(2, 2, 2)
plt.contour(X, Y, Z, 20)

plt.subplot(2, 2, 3)
plt.contourf(X, Y, Z)

plt.subplot(2, 2, 4)
plt.contourf(X, Y, Z, 20)

plt.tight_layout()
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.contourf(X, Y, Z, 20, cmap='viridis')
plt.colorbar()
plt.title('Viridis')

plt.subplot(2, 2, 2)
plt.contourf(X, Y, Z, 20, cmap='gray')
plt.colorbar()
plt.title('Gray')

plt.subplot(2, 2, 3)
plt.contourf(X, Y, Z, 20, cmap='autumn')
plt.colorbar()
plt.title('Autumn')

plt.subplot(2, 2, 4)
plt.contourf(X, Y, Z, 20, cmap='winter')
plt.colorbar()
plt.title('Winter')

plt.tight_layout()
plt.show()
```

```python
auto = pd.read_csv('auto-mpg.csv')
auto.head()
```

```python
plt.hist2d('hp', 'mpg', data=auto, bins=(20, 20), range=((40, 235), (8, 48)))
plt.colorbar()
plt.xlabel('Horse power [hp]')
plt.ylabel('Miles per gallon [mpg]')
plt.title('hist2d() plot')
plt.show()
```

```python
plt.hexbin('hp', 'mpg', data=auto, gridsize=(15, 12), extent=(40, 235, 8, 48))
plt.colorbar()
plt.xlabel('Horse power [hp]')
plt.ylabel('Miles per gallon [mpg]')
plt.title('hexbin() plot')
plt.show()
```
