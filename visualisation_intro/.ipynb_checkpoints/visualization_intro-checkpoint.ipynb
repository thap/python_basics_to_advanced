```python
u = np.linspace(-2, 2, 41)
v = np.linspace(-1, 1, 21)

X, Y = np.meshgrid(u, v)
Z = np.sin(3*np.sqrt(X**2 + Y**2))

plt.pcolor(Z)
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.contour(X, Y, Z)

plt.subplot(2, 2, 2)
plt.contour(X, Y, Z, 20)

plt.subplot(2, 2, 3)
plt.contourf(X, Y, Z)

plt.subplot(2, 2, 4)
plt.contourf(X, Y, Z, 20)

plt.tight_layout()
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.contourf(X, Y, Z, 20, cmap='viridis')
plt.colorbar()
plt.title('Viridis')

plt.subplot(2, 2, 2)
plt.contourf(X, Y, Z, 20, cmap='gray')
plt.colorbar()
plt.title('Gray')

plt.subplot(2, 2, 3)
plt.contourf(X, Y, Z, 20, cmap='autumn')
plt.colorbar()
plt.title('Autumn')

plt.subplot(2, 2, 4)
plt.contourf(X, Y, Z, 20, cmap='winter')
plt.colorbar()
plt.title('Winter')

plt.tight_layout()
plt.show()
```

```python
image = plt.imread('640px-Unequalized_Hawkes_Bay_NZ.jpg')
pmin, pmax = image.min(), image.max()
print('The smallest and largest pixel intensities are {} and {}'.format(pmin, pmax))
```

```python
rescaled_image = 256*(image-pmin)/(pmax-pmin)
print('The rescaled smallest and largest pixel intensities are {} and {}'
      .format(rescaled_image.min(), rescaled_image.max()))
```

```python
plt.subplot(1, 2, 1)
plt.title('original image')
plt.axis('off')
plt.imshow(image)

plt.subplot(1, 2, 2)
plt.title('rescaled image')
plt.axis('off')
plt.imshow(rescaled_image)

plt.show()
```

```python
stocks = pd.read_csv('stocks.csv', parse_dates=True, index_col='Date')
stocks.head()
```

```python
stocks.plot()
plt.legend(loc='upper left')
plt.xticks(rotation=60)
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.xticks(rotation=45)
plt.title('AAPL: 2001 to 2011')
plt.plot(stocks.AAPL, color='blue')

view = stocks.AAPL['2007':'2008']
plt.subplot(2, 1, 2)
plt.xticks(rotation=45)
plt.title('APPL: 2007 to 2008')
plt.plot(view, color='black')
plt.tight_layout()
plt.show()
```

```python
view = stocks.AAPL['2007-11':'2008-04']
plt.plot(stocks.AAPL, color='blue')
plt.xticks(rotation=45)
plt.title('AAPL: 2001-2011')
plt.axes([0.25, 0.5, 0.35, 0.35])
plt.plot(view, color='red')
plt.xticks(rotation=45)
plt.title('2007/11-2008/04')
plt.tight_layout()
plt.show()
```

```python
mean_30 = stocks.AAPL.rolling(window=30).mean()
mean_75 = stocks.AAPL.rolling(window=75).mean()
mean_125 = stocks.AAPL.rolling(window=125).mean()
mean_250 = stocks.AAPL.rolling(window=250).mean()
```

```python
plt.subplot(2, 2, 1)
plt.plot(mean_30, color='green')
plt.xticks(rotation=60)
plt.title('30d averages')

plt.subplot(2, 2, 2)
plt.plot(mean_75, 'red')
plt.xticks(rotation=60)
plt.title('75d averages')

plt.subplot(2, 2, 3)
plt.plot(mean_125, 'blue')
plt.xticks(rotation=60)
plt.title('125d averages')

plt.subplot(2, 2, 4)
plt.plot(mean_250, 'cyan')
plt.xticks(rotation=60)
plt.title('250d averages')

plt.tight_layout()
plt.show()
```

```python
std_30 = stocks.AAPL.rolling(window=30).std()
std_75 = stocks.AAPL.rolling(window=75).std()
std_125 = stocks.AAPL.rolling(window=125).std()
std_250 = stocks.AAPL.rolling(window=250).std()
```

```python
plt.plot(std_30, color='green', label='30d')
plt.plot(std_75, color='red', label='75d')
plt.plot(std_125, color='blue', label='125d')
plt.plot(std_250, color='cyan', label='250d')
plt.legend(loc='upper left')
plt.title('Moving standard deviations')
plt.show()
```

```python
image = plt.imread('640px-Unequalized_Hawkes_Bay_NZ.jpg')
plt.subplot(1, 2, 1)
plt.title('Original image')
plt.axis('off')
plt.imshow(image, cmap='gray')

pixels = image.flatten()
plt.subplot(1, 2, 2)
plt.xlim((0, 255))
plt.title('Normalized histogram')
plt.hist(pixels, bins=64, range=(0, 256), density=True, color='red', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
image = plt.imread('640px-Unequalized_Hawkes_Bay_NZ.jpg')
plt.subplot(1, 2, 1)
plt.title('Original image')
plt.axis('off')
plt.imshow(image, cmap='gray')

pixels = image.flatten()
plt.subplot(1, 2, 2)
pdf = plt.hist(pixels, bins=64, range=(0, 256), density=True, color='red', alpha=0.4)
plt.grid('off')
plt.twinx()
cdf = plt.hist(pixels, bins=64, range=(0, 256), density=True, cumulative=True, color='blue', alpha=0.4)
plt.xlim((0, 256))
plt.grid('off')
plt.title('PDF and CDF of the original image')

plt.tight_layout()
plt.show()
```

```python
image = plt.imread('640px-Unequalized_Hawkes_Bay_NZ.jpg')

pixels = image.flatten()

cdf, bins, patches = plt.hist(pixels, bins=256, range=(0, 256), density=True, \
                              cumulative=True)
new_pixels = np.interp(pixels, bins[:-1], cdf * 255)
new_image = new_pixels.reshape(image.shape)

plt.subplot(1, 2, 1)
plt.title('Equalized image')
plt.axis('off')
plt.imshow(new_image, cmap='gray')

plt.subplot(1, 2, 2)
pdf = plt.hist(new_pixels, bins=64, range=(0, 256), density=False, color='red', \
               alpha=0.4)
plt.grid('off')

plt.twinx()
plt.xlim((0, 256))
plt.grid('off')

plt.title('PDF and CDF of equalized image')

cdf = plt.hist(new_pixels, bins=64, range=(0, 256), cumulative=True, density=True, \
               color='blue', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
image = plt.imread('hs-2004-32-b-small_web.jpg')

red, green, blue = image[:,:,0], image[:,:,1], image[:,:,2]
rpixels = red.flatten()
bpixels = blue.flatten()
gpixels = green.flatten()

plt.subplot(1, 2, 1)
plt.title('Original image')
plt.axis('off')
plt.imshow(image)

plt.subplot(1, 2, 2)
plt.title('Histograms from color image')
plt.xlim((0, 256))
plt.hist(rpixels, bins=64, density=True, color='red', alpha=0.2)
plt.hist(bpixels, bins=64, density=True, color='blue', alpha=0.2)
plt.hist(gpixels, bins=54, density=True, color='green', alpha=0.2)

plt.tight_layout()
plt.show()
```

```python
image = plt.imread('hs-2004-32-b-small_web.jpg')

red, green, blue = image[:,:,0], image[:,:,1], image[:,:,2]
rpixels = red.flatten()
bpixels = blue.flatten()
gpixels = green.flatten()

plt.subplot(2, 2, 1)
plt.grid('off')
plt.xticks(rotation=60)
plt.xlabel('red')
plt.ylabel('green')
plt.hist2d(x=rpixels, y=gpixels, bins=(32, 32))

plt.subplot(2, 2, 2)
plt.grid('off')
plt.xticks(rotation=60)
plt.xlabel('green')
plt.ylabel('blue')
plt.hist2d(x=gpixels, y=bpixels, bins=(32, 32))

plt.subplot(2, 2, 3)
plt.grid('off')
plt.xticks(rotation=60)
plt.xlabel('blue')
plt.ylabel('red')
plt.hist2d(x=bpixels, y=rpixels, bins=(32, 32))

plt.tight_layout()
plt.show()
```
