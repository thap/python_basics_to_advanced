```python
editions = pd.read_csv('Summer Olympic medals/Summer Olympic medalists 1896 to 2008 - EDITIONS.tsv', sep='\t')
edi = editions[['Edition', 'Grand Total', 'City', 'Country']]
edi
```

```python
ioc_codes = pd.read_csv('Summer Olympic medals/Summer Olympic medalists 1896 to 2008 - IOC COUNTRY CODES.csv')
ioc = ioc_codes[['Country', 'NOC']]
ioc.head()
```

```python
ioc.tail()
```

```python
medals = pd.read_csv('Summer Olympic medals/Summer Olympic medalists 1896 to 2008 - ALL MEDALISTS.tsv', header=4, sep='\t')
medals.head()
```

```python
medals.tail()
```

```python
medal_counts = medals.pivot_table(aggfunc='count', index='Edition', values='Athlete', columns='NOC')
medal_counts.head()
```

```python
medal_counts.tail()
```

```python
totals = edi.set_index('Edition')
tot = totals['Grand Total']
tot
```

```python
fractions = medal_counts.divide(tot, axis='rows')
fractions.head()
```

```python
fractions.tail()
```

```python
mean_fractions = fractions.expanding().mean()
fractions_change = mean_fractions.pct_change()*100
fractions_change_noidx = fractions_change.reset_index()
fractions_change_noidx.head()
```

```python
fractions_change_noidx.tail()
```

```python
hosts = pd.merge(edi, ioc, how='left')
hosts.head()
```

```python
ho = hosts[['Edition', 'NOC']].set_index('Edition')
print(ho.loc[ho.NOC.isnull()])
```

```python
ho.loc[1972, 'NOC'] = 'FRG'
ho.loc[1980, 'NOC'] = 'URS'
ho.loc[1988, 'NOC'] = 'KOR'
print(ho.loc[ho.NOC.isnull()])
```

```python
fractions_change_noidx.info()
```

```python
reshaped = pd.melt(fractions_change_noidx, id_vars='Edition', value_name='Change')
reshaped.shape
```

```python
chn = reshaped[reshaped['NOC'] == 'CHN']
chn.tail()
```

```python
merged = pd.merge(reshaped, hosts)
print(merged.head())
```

```python
influence = merged.set_index('Edition').sort_index()
influence.head()
```

```python
change = influence['Change']
ax = change.plot(kind='bar')
ax.set_ylabel('% Change of Host Country Medal Count')
ax.set_title('Is there a Host Country Advantage?')
ax.set_xticklabels(edi['City'])
plt.show()
```
