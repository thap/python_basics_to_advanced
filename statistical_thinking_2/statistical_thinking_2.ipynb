```python
def ecdf(data):
    return np.sort(data), np.arange(1, len(data)+1) / len(data)
```

```python
x, y = ecdf(nohitter_times)
x_, y_ = ecdf(inter_nohitter_time)
plt.plot(x_, y_)
plt.plot(x, y, marker='.', linestyle='none')
plt.margins(0.02)
plt.xlabel('Games between no-hitters')
plt.ylabel('CDF')
plt.show()
```

```python
import pandas as pd

female = pd.read_csv('female_literacy_fertility.csv')
female
```

```python
illiteracy = 100 - female['female literacy']
plt.plot(illiteracy, female.fertility, marker='.', linestyle='none')
plt.margins(0.02)
plt.xlabel('percent illiterate')
plt.ylabel('fertility')
plt.show()
```

```python
from scipy.stats import pearsonr

print(pearsonr(illiteracy, female.fertility)[0])
```

```python
plt.plot(illiteracy, female.fertility, marker='.', linestyle='none')
plt.margins(0.02)
plt.xlabel('percent illiterate')
plt.ylabel('fertility')
a, b = np.polyfit(illiteracy, female.fertility, 1)
print('slope =', a, 'children per woman / percent illiterate')
print('intercept =', b, 'children per woman')
x = np.array([0, 100])
y = a * x + b
plt.plot(x, y)
plt.show()
```

```python
a_ = np.linspace(0, 0.1, 200)
rss = np.empty_like(a_)
for i, a in enumerate(a_):
    rss[i] = np.sum((female.fertility - a * illiteracy - b)**2)
    
plt.plot(a_, rss, '-')
plt.xlabel('slope (children per woman/percent illiterate)')
plt.ylabel('sum of square of residuals')
plt.show()
```
